# 7. Filter for employees with `PercentSalaryHike > 15 **AND** `PerformanceRating == 4.

```python
filter_emp = ((df['PercentSalaryHike'] > 15) & (df['PerformanceRating'] == 4))
filter_emp
```

# 8. Create a pivot table showing the **average** `MonthlyIncome
## for each `Department (rows) broken down by JobSatisfaction (columns).

```python
pivot_table = df.pivot_table(
    values='MonthlyIncome',
    index='Department',
    columns='JobSatisfaction',
    aggfunc='mean'
)
pivot_table
```

# 9. Create a new column
## Seniority  If YearsAtCompany > 10, value is'Senior' Otherwise, value is'Junior'

```python
df['Seniority'] = np.where(df['YearsAtCompany'] > 10, 'Senior', 'Junior')
df['Seniority'].value_counts()
```

```python
df[['YearsAtCompany','Seniority']]
```

# 10. Identify 'High Risk' employees who meet **ALL** these criteria:
**1.MaritalStatus is 'Single'**
**2. JobSatisfaction is less than**
**3.DistanceFromHome > 10**
**Display their EmployeeID and `JobRole**

```python
high_risk = df[
    (df['MaritalStatus'] == 'Single') &
    (df['JobSatisfaction'] < 3) &
    (df['DistanceFromHome'] > 10) ]

high_risk[['EmployeeID', 'JobRole']]
```
